```python
def drawGraph7():
    plt.figure(figsize=(14,10))
    plt.scatter(data_result["인구수"],data_result["총계"],s=50)
    plt.xlabel("인구수")
    plt.ylabel("CCTV")
    plt.grid()
    plt.show()
```

```python
drawGraph7()
```

```python
fp1=np.polyfit(data_result["인구수"],data_result["총계"],1)
fp1
```

```python
f1=np.poly1d(fp1)
```

```python
fx=np.linspace(100000, 700000,100)  #등간격으로 데이터 생성
```

```python
def drawGraph8():
    plt.figure(figsize=(14,10))
    plt.scatter(data_result["인구수"],data_result["총계"],s=50)
    plt.plot(fx,f1(fx), ls='dashed',lw=3, color='g')
    plt.xlabel("인구수")
    plt.ylabel("CCTV")
    plt.grid()
    plt.show()
```

```python
drawGraph8()
```

```python
data_result["오차"]=data_result["총계"]-f1(data_result["인구수"])

df_descending=data_result.sort_values(by="오차",ascending=False)
df_ascending=data_result.sort_values(by="오차",ascending=True)
```

```python
df_descending.head()
```

```python
df_ascending.head()
```

```python
color_step=["#e74c3c","#2ecc71","#95a5a6","#2ecc71","#3498db","#3498db"]
my_colormap=ListedColormap(color_step)
```

```python
def drawGraph9():
    plt.figure(figsize=(14,10))
    plt.scatter(data_result["인구수"], data_result["총계"],c=data_result["오차"],s=50,cmap=my_colormap)
    plt.plot(fx,f1(fx),ls='dashed',lw=3,color='gray')

    for n in range(5):
        plt.text(df_descending["인구수"][n]*1.02,
                 df_descending["총계"][n]*0.98,
                 df_descending.index[n],fontsize=15)
        
        plt.text(df_ascending["인구수"][n]*1.02,
                 df_ascending["총계"][n]*0.98,
                 df_ascending.index[n],fontsize=15)
    plt.xlabel("인구수")
    plt.ylabel("인구수")
    plt.colorbar()
    plt.grid()
    
                 
```

```python
drawGraph9()
```
