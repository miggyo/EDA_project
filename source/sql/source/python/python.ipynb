```python
conn = mysql.connector.connect(
    host = 'database-1.cb2iuock61zt.ap-northeast-2.rds.amazonaws.com',
    port = 3306,
    user =  'root',
    password = 'asdf1234',
    database = 'amrdb'
)
```

```python
df= pd.read_csv('../../data/2020_crime.csv', encoding='euc-kr')
df.head()
```

```python
sql = """insert into crime_status values ("2020",%s,%s,%s,%s)""" #따옴표를 세 개씩 (""") 사용하는 것은 여러 줄에 걸친 문자열을 정의하는 Python의 표준 문법
cursor = conn.cursor(buffered=True)
```

```python
df.iterrows()
```

```python
for i , row in df.iterrows(): #iterrows()는 DataFrame의 각 행을 (인덱스, 행 데이터)의 튜플로 반환
    cursor.execute(sql, tuple(row)) # execute 메서드를 호출하여 SQL 쿼리를 실행하고, 튜플의 값이 쿼리에 매개변수로 전달되어 데이터베이스에 레코드를 추가
    print(tuple(row))
    conn.commit()
```
