```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
x= np.linspace(0,14,100)
y1=np.sin(x)
y2=2*np.sin(x+0.5)
y3=3*np.sin(x+1.0)
y4=4*np.sin(x+1.5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
sns.despine();plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
sns.despine(offset=10)
plt.show()
```

```python
tips=sns.load_dataset("tips")
tips.head(5)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day",y="total_bill",data=tips,palette="Set3")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day",y="total_bill",hue="smoker",data=tips,palette="Set3")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day",y="total_bill",data=tips,palette="Set3")
sns.swarmplot(x="day",y="total_bill", data=tips,color=".25")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=tips,height=7)
plt.show()
```

```python
flights=sns.load_dataset("flights")
flights.head(5)
```

```python
#pivot : data 형태를 바꿔줌 새로운 기준으로 정리해줌
```

```python
flights = flights.pivot(index="month", columns="year", values="passengers")
flights.head(5)
```

```python
#heatmap colormap 같은것 annotation : true-숫자보임,false-숫자 안보임 fmat : d-정수형태 cmap : 원하는 색조합으로 생성가능
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights,annot=True,fmt="d")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights,annot=True,fmt="d",cmap="YlGnBu")
plt.show()
```
